### Which should I use?

* When plotting something quickly, use pyplot method.
* When plotting more advanced, use the Matplotlib OO method.

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
over_50
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

# ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
# Set the style 
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap='winter')

ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```
